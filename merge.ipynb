```python
df1['FLOOD_CAUSE'].value_counts()
```

```python
df1['FLOOD_CAUSE']= df1['FLOOD_CAUSE'].str.replace('Planned Dam Release', 'Dam')
df1['FLOOD_CAUSE'].value_counts()
```

```python
df1.info()
```

```python
df1['FLOOD_IMPACT_IN_DAYS'].value_counts()
```
